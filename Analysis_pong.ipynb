```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Plot the results of the first training session
## Stochastic policy

df = pd.read_excel(r'/content/gdrive/MyDrive/Excelsheets/Pong_results_colab_total1.xlsx')
# print(df.head())  # print first 5 rows of the dataframe

# Plot episode and time (x-axis) against total_reward (y-axis)
df['time'] = df['time'] / 3600
# df = df[:8000]


ax = plt.gca()
fig1 = df.plot(kind='line',x='episode',y=['reward_per_game', 'running_mean'], ax=ax)
fig1.set(xlabel="Episode", ylabel="Reward")

# fig1 = df.plot(kind='line',x='time',y=['reward_per_game', 'running_mean'], ax=ax)
# fig1.set(xlabel="Time [h]", ylabel="Reward")

# a_b = np.polyfit(df['episode'],df['running_mean'],1)
# fit = np.poly1d(a_b)
# df.insert(5,'Regression_line',fit(df['episode']))
# fig2 = df.plot(x = 'episode',y='running_mean')
# fig2 = df.plot(x='episode', y='Regression_line',color='Red',ax=fig2)
# fig2.set(xlabel="Episode", ylabel="Reward")

a_b = np.polyfit(df['time'],df['running_mean'],1)
fit = np.poly1d(a_b)
df.insert(5,'Regression_line',fit(df['time']))
fig2 = df.plot(x = 'time',y='running_mean')
fig2 = df.plot(x='time', y='Regression_line',color='Red',ax=fig2)
fig2.set(xlabel="Episode", ylabel="Reward")

print(a_b)

plt.show()

```
